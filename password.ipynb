```python
# password generator
passwd = ''

def generate_random_password(set):
    import random
    # shuffling given character set
    random.shuffle(set)
    
    # picking random characters from the list as many as defined in length
    password = []
    for i in range(length):
        password.append(random.choice(set))
    
    # shuffling the picked characters
    random.shuffle(password)
    
    # convert to string and return
    pwstr = ''.join([str(elem) for elem in password])
    return pwstr


# set parameters for password.
length = 2
charset = alphabets # lower case letters
passwd = generate_random_password(charset)
print(passwd)
```

```python
# using brute force method for cracking
from random import * # pseudorandom

# initializing of variables
guess = ''
tries = 0 # no functional purpose but to visualize guess count

# timetracking
start = time.time()

# using while loop to generate random strings until one of them matches passwd
while (guess != passwd):
    guess = ''
    tries = tries+1
    # generating random passwords using for loop
    for letter in range(len(passwd)):
        guess_letter = alphabets[randint(0, 25)] # alphabets with 26 entries
        guess = str(guess_letter) + str(guess)
    
# calculate how long it took
duration = (time.time() - start)
    
# printing the matched password and duration
print('The password '+guess+' was found in ' + str(tries) + ' tries')
print('It took '+str(duration) + ' seconds')
```

```python
### HASH and dictionary search###
import hashlib
 
def hash(password):
    result = hashlib.sha256(password.encode()) # using sha256 here
    return result.hexdigest()


# passwd = 'From_Trust_To_BlockChain_2022' # also experiment with a unique password
passwd = 'password1'
passwd_hash = hash(passwd)
print(passwd_hash)
```

```python
# read in the leaked password list. define your own path for it, or use something like urllib for remote download
with open('leaked_passwords.txt') as f:
    data = [line.rstrip() for line in f]


# bruteforce dictionary search function    
def bruteforce():
    # apply same hash for each line in the file and compare to passwd_hash
    for password_in_file in data:
        if hash(password_in_file) == passwd_hash:
            return password_in_file # return the password when match is found


# Run the attack
guess_pw = bruteforce()
if guess_pw is None:
    print('Your password was not found, you are safe!')
else:
    print('Your password is \"'+guess_pw+'\" has been compromised. Set a new password now.')
```
